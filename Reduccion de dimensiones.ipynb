# Reduccion de dimensiones

**Algoritmo PCA:**

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
from sklearn import decomposition
```

```
X1,Y1 = make_blobs(n_features=10, n_samples=100,centers=4, random_state=4, cluster_std=1)
print(X1.shape)
print(Y1.shape)
```

```
matrix=X1
df=pd.DataFrame(matrix,index=range(matrix.shape[0]),columns=range(matrix.shape[1]))
print(len(df.columns))
```

```
df['y']=Y1
print(len(df.columns))
```

```
df.rename(columns={0:'x1',
                   1:'x2',
                   2:'x3',
                   3:'x4',
                   4:'x5',
                   5:'x6',
                   6:'x7',
                   7:'x8',
                   8:'x9',
                   9:'x10'
                   }, inplace=True)
print(df.columns)
```

```
sns.lmplot(x='x2', y='x3', data=df, fit_reg=False, hue='y', legend=True, scatter_kws={"s":80})
pca=decomposition.PCA(n_components=4)
pc=pca.fit_transform(X1)
```

```
df=pd.DataFrame({'var':pca.explained_variance_ratio_, 'PC':['PC1','PC2','PC3','PC4']})
print(df)
```

```
plt.figure()
sns.barplot(x='PC',y='var',data=df, color='c')
```

```
pc_df=pd.DataFrame(data=pc, columns=['PC1','PC2','PC3','PC4'])
pc_df['Cluster']=Y1
sns.lmplot(x="PC1", y="PC2",
           data=pc_df,
           fit_reg=False,
           hue='Cluster',
           scatter_kws={"s":80})
plt.show()
```
